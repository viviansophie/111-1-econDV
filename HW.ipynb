```r
# load package and data
library(ggplot2)
library()
```

```r
mos <- read.csv("mos_test.csv")
# data(mos, package="ggplot2")
head(mos)
mos_select <- mos[mos$Scenario	%in% c("BL1", "PL1", "P1","L1"), ]
head(mos_select)

# Scatterplot
theme_set(theme_bw())  # pre-set the bw theme.
g <- ggplot(mos_select, aes(Score, op)) + 
  labs(subtitle="mos:test",
       title="Bubble chart")

g + geom_jitter(aes(col=Scenario, size= graf)) + 
  geom_smooth(aes(col=Scenario), method="lm", se=F)

# graf ->畫質
# inter -> 互動性
# op -> 意見
```
